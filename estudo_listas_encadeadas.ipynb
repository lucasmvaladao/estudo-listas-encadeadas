#**Cap√≠tulo 1 - Introdu√ß√£o √†s Listas Encadeadas**



---



üîó O que s√£o listas encadeadas?

As listas encadeadas s√£o estruturas de dados lineares que armazenam elementos de forma sequencial, mas com uma diferen√ßa importante em rela√ß√£o √†s listas comuns do Python (list[]): os elementos n√£o ocupam posi√ß√µes consecutivas na mem√≥ria.

Em vez disso, cada elemento ‚Äî chamado de n√≥ ‚Äî guarda duas informa√ß√µes principais:

* Um valor (ou dado);

* Um ponteiro (tamb√©m chamado de "refer√™ncia") para o pr√≥ximo n√≥ da lista.


üí° Imagine uma fila de caixas conectadas por cordas: Cada caixa cont√©m um item (o dado) e uma corda que aponta para a pr√≥xima caixa. Essa √© a ideia por tr√°s das listas encadeadas!



---



üì¶ Tipos de listas encadeadas
Existem dois tipos principais:

‚û§ Lista Encadeada Simples
* Cada n√≥ armazena um valor e um ponteiro para o pr√≥ximo n√≥.

* A navega√ß√£o acontece em uma √∫nica dire√ß√£o: do in√≠cio at√© o fim da lista.

‚û§ Lista Encadeada Dupla
Cada n√≥ armazena:

* Um valor;

* Um ponteiro para o pr√≥ximo n√≥;

* E um ponteiro para o n√≥ anterior.

Isso permite navegar nos dois sentidos: tanto do in√≠cio para o fim, quanto do fim para o in√≠cio.

```python
# Fun√ß√£o para criar um n√≥ da lista duplamente encadeada
def criar_no(valor, anterior=None, proximo=None):
    return {"valor": valor, "anterior": anterior, "proximo": proximo}

# Criando e encadeando os n√≥s do fim para o in√≠cio
no_final = criar_no(30)

no_meio = criar_no(20, proximo=no_final)
no_final["anterior"] = no_meio
no_inicio = criar_no(10, proximo=no_meio)
no_meio["anterior"] = no_inicio

# Fun√ß√£o para imprimir a lista do in√≠cio ao fim
def imprimir_lista(inicio):
    atual = inicio
    while atual is not None:
        print(atual["valor"])
        atual = atual["proximo"]

# Fun√ß√£o para imprimir a lista do fim ao in√≠cio
def imprimir_lista_reversa(fim):
    atual = fim
    while atual is not None:
        print(atual["valor"])
        atual = atual["anterior"]

# Testando
print("Ordem normal:")
imprimir_lista(no_inicio)

print("\nOrdem reversa:")
imprimir_lista_reversa(no_final)
```

## ‚úÖ INSERIR ORDENADO

üí° Simplesmente Encadeada

```python
def inserir_ordenado_simples(lista, valor):
    novo_no = {"valor": valor, "proximo": None}
    if lista is None or valor < lista["valor"]:
        novo_no["proximo"] = lista
        return novo_no
    atual = lista
    while atual["proximo"] is not None and atual["proximo"]["valor"] < valor:
        atual = atual["proximo"]
    novo_no["proximo"] = atual["proximo"]
    atual["proximo"] = novo_no
    return lista

def imprimir_lista(inicio):
    atual = inicio
    while atual is not None:
        print(atual["valor"])
        atual = atual["proximo"]

lista_simples = None
lista_simples = inserir_ordenado_simples(lista_simples, 20)
lista_simples = inserir_ordenado_simples(lista_simples, 10)
lista_simples = inserir_ordenado_simples(lista_simples, 30)

imprimir_lista(lista_simples)
```

üîÅ Duplamente Encadeada

```python
# Fun√ß√£o para criar um n√≥ da lista duplamente encadeada
def criar_no_duplo(valor, anterior=None, proximo=None):
    return {"valor": valor, "anterior": anterior, "proximo": proximo}

def inserir_ordenado_duplo(lista, valor):
    novo_no = criar_no_duplo(valor)
    if lista is None or valor < lista["valor"]:
        novo_no["proximo"] = lista
        if lista is not None:
            lista["anterior"] = novo_no
        return novo_no
    atual = lista
    while atual["proximo"] is not None and atual["proximo"]["valor"] < valor:
        atual = atual["proximo"]
    novo_no["proximo"] = atual["proximo"]
    novo_no["anterior"] = atual
    if atual["proximo"] is not None:
        atual["proximo"]["anterior"] = novo_no
    atual["proximo"] = novo_no
    return lista

def imprimir_lista(inicio):
    atual = inicio
    while atual is not None:
        print(atual["valor"])
        atual = atual["proximo"]

lista_dupla = None
lista_dupla = inserir_ordenado_duplo(lista_dupla, 20)
lista_dupla = inserir_ordenado_duplo(lista_dupla, 10)
lista_dupla = inserir_ordenado_duplo(lista_dupla, 30)

imprimir_lista(lista_dupla)
```

# **Cap√≠tulo 4 - Remo√ß√£o de Elementos (In√≠cio, Meio e Fim)**

## ‚úÖ REMO√á√ÉO DO IN√çCIO

üí° Simplesmente Encadeada

```python
# Fun√ß√£o para criar um n√≥ da lista
def criar_no(valor, proximo=None):
    return {"valor": valor, "proximo": proximo}

def remover_inicio_simples(lista):
    # Se a lista n√£o for vazia, retorna o segundo n√≥
    return lista["proximo"] if lista else None

def imprimir_lista(inicio):
    # Imprime todos os elementos da lista
    atual = inicio
    while atual:
        print(atual["valor"])
        atual = atual["proximo"]

# Criando os n√≥s
no_final = criar_no(30)  # O √∫ltimo n√≥
no_meio = criar_no(20, no_final)  # N√≥ do meio
no_inicio = criar_no(10, no_meio)  # O primeiro n√≥ (in√≠cio da lista)

# Agora, a lista come√ßa em no_inicio
lista_simples = no_inicio


# Removendo o primeiro elemento
lista_simples = remover_inicio_simples(lista_simples)

# Imprimindo a lista ap√≥s remover o in√≠cio
imprimir_lista(lista_simples)
```

üîÅ Duplamente Encadeada

```python
# Fun√ß√£o para criar um n√≥ da lista duplamente encadeada
def criar_no_duplo(valor, anterior=None, proximo=None):
    return {"valor": valor, "anterior": anterior, "proximo": proximo}

# Fun√ß√£o para remover o primeiro n√≥ da lista duplamente ligada
def remover_inicio_duplo(lista):
    if lista is None:
        return None  # Se a lista estiver vazia, retorna None

    proximo = lista["proximo"]  # Guarda o segundo n√≥
    if proximo:
        proximo["anterior"] = None  # Remove a refer√™ncia ao n√≥ anterior

    return proximo  # Retorna o segundo n√≥, que agora √© o "in√≠cio" da lista

# Fun√ß√£o para imprimir a lista
def imprimir_lista(inicio):
    atual = inicio
    while atual:
        print(atual["valor"])
        atual = atual["proximo"]

# Criando a lista duplamente ligada
no_final = criar_no_duplo(30)
no_meio = criar_no_duplo(20, None, no_final)
no_final["anterior"] = no_meio
no_inicio = criar_no_duplo(10, None, no_meio)
no_meio["anterior"] = no_inicio

# A lista come√ßa em no_inicio
lista_dupla = no_inicio


# Removendo o primeiro n√≥ (no_inicio)
lista_dupla = remover_inicio_duplo(lista_dupla)

# Imprimindo a lista ap√≥s remover o in√≠cio
imprimir_lista(lista_dupla)
```

## ‚úÖ REMO√á√ÉO DO FIM

üí° Simplesmente Encadeada

```python
# Fun√ß√£o para criar um n√≥ da lista
def criar_no(valor, proximo=None):
    return {"valor": valor, "proximo": proximo}

# Fun√ß√£o para remover o √∫ltimo n√≥ de uma lista simplesmente ligada
def remover_fim_simples(lista):
    if lista is None or lista["proximo"] is None:
        return None  # Lista vazia ou com 1 s√≥ elemento

    atual = lista
    # Vai at√© o pen√∫ltimo n√≥
    while atual["proximo"] and atual["proximo"]["proximo"] is not None:
        atual = atual["proximo"]

    atual["proximo"] = None  # Remove o √∫ltimo n√≥
    return lista

# Fun√ß√£o para imprimir a lista
def imprimir_lista(inicio):
    atual = inicio
    while atual:
        print(atual["valor"])
        atual = atual["proximo"]

# Criando a lista
no3 = criar_no(30)
no2 = criar_no(20, no3)
no1 = criar_no(10, no2)

# Lista come√ßa com o n√≥ 'no1'
lista_simples = no1

# Removendo o √∫ltimo n√≥
lista_simples = remover_fim_simples(lista_simples)

# Imprimindo a lista depois de remover o √∫ltimo n√≥
imprimir_lista(lista_simples)
```

üîÅ Duplamente Encadeada

```python
# Fun√ß√£o para criar um n√≥ da lista duplamente encadeada
def criar_no_duplo(valor, anterior=None, proximo=None):
    return {"valor": valor, "anterior": anterior, "proximo": proximo}

def imprimir_lista(inicio):
    atual = inicio
    while atual:
        print(atual["valor"])
        atual = atual["proximo"]

def remover_fim_duplo(lista):
    if lista is None or lista["proximo"] is None:
        return None

    atual = lista
    while atual["proximo"]["proximo"] is not None:
        atual = atual["proximo"]

    atual["proximo"]["anterior"] = None  # Remove refer√™ncia ao pen√∫ltimo
    atual["proximo"] = None              # Remove o √∫ltimo
    return lista

# Criando lista duplamente ligada: 10 <-> 20 <-> 30
no_final = criar_no_duplo(30)
no_meio = criar_no_duplo(20, None, no_final)
no_final["anterior"] = no_meio
no_inicio = criar_no_duplo(10, None, no_meio)
no_meio["anterior"] = no_inicio

lista_dupla = no_inicio

# Removendo o √∫ltimo
lista_dupla = remover_fim_duplo(lista_dupla)


imprimir_lista(lista_dupla)
```

## ‚úÖ REMO√á√ÉO POR VALOR

üí° Simplesmente Encadeada

```python
# Fun√ß√£o para criar um n√≥ da lista
def criar_no(valor, proximo=None):
    return {"valor": valor, "proximo": proximo}

def imprimir_lista(inicio):
    atual = inicio
    while atual:
        print(atual["valor"])

        atual = atual["proximo"]

def remover_valor_simples(lista, valor):
    if lista is None:
        return None

    if lista["valor"] == valor:
        return lista["proximo"]  # Remove o primeiro n√≥ se for o alvo

    atual = lista

    while atual["proximo"] and atual["proximo"]["valor"] != valor: #Enquanto houver um pr√≥ximo n√≥ e o valor do pr√≥ximo n√£o for o valor que queremos remover... continue andando.
        atual = atual["proximo"]  # Procura o valor

    if atual["proximo"]: # Isso verifica se existe um pr√≥ximo n√≥. Se sim, podemos acessar ele.
        atual["proximo"] = atual["proximo"]["proximo"]  # Pula o n√≥ com o valor
    return lista

# Criando a lista: 10 ‚Üí 20 ‚Üí 30 ‚Üí 40
no4 = criar_no(40)
no3 = criar_no(30, no4)
no2 = criar_no(20, no3)
no1 = criar_no(10, no2)

lista = no1


# Remover o valor 30
lista = remover_valor_simples(lista, 30)

print("\n‚ùå Ap√≥s remover o valor 30:")
imprimir_lista(lista)


# Remover um valor que n√£o existe (ex: 99)
lista = remover_valor_simples(lista, 99)
print("\n‚ÑπÔ∏è Ap√≥s tentar remover o valor 99 (inexistente):")
imprimir_lista(lista)
```

üîÅ Duplamente Encadeada

```python
# Fun√ß√£o para criar um n√≥ da lista duplamente encadeada
def criar_no_duplo(valor, anterior=None, proximo=None):
    return {"valor": valor, "anterior": anterior, "proximo": proximo}

# Fun√ß√£o para imprimir a lista
def imprimir_lista(inicio):
    atual = inicio
    while atual:
        print(atual["valor"])
        atual = atual["proximo"]

# Fun√ß√£o para remover um valor espec√≠fico da lista duplamente ligada
def remover_valor_duplo(lista, valor):
    if lista is None:
        return None

    # Caso o valor esteja no primeiro n√≥
    if lista["valor"] == valor:
        proximo = lista["proximo"]
        if proximo:
            proximo["anterior"] = None  # limpa o anterior
        return proximo

    # Procurando o valor na lista
    atual = lista
    while atual and atual["valor"] != valor:
        atual = atual["proximo"]

    # Se encontrou, faz os ajustes nos ponteiros
    if atual:
        if atual["anterior"]:
            atual["anterior"]["proximo"] = atual["proximo"]
        if atual["proximo"]:
            atual["proximo"]["anterior"] = atual["anterior"]

    return lista

# Criando a lista: 100 <-> 200 <-> 300 <-> 400
no4 = criar_no_duplo(400)
no3 = criar_no_duplo(300, no4)
no2 = criar_no_duplo(200, no3)
no1 = criar_no_duplo(100, no2)

# Ajustando os ponteiros "anterior"
no2["anterior"] = no1
no3["anterior"] = no2
no4["anterior"] = no3

# A lista come√ßa em no1
lista = no1


# Remover valor do meio (300)
lista = remover_valor_duplo(lista, 300)
print("\nAp√≥s remover o valor 300:")
imprimir_lista(lista)

# Tentar remover valor inexistente (999)
lista = remover_valor_duplo(lista, 999)
print("\nAp√≥s tentar remover valor 999 (inexistente):")
imprimir_lista(lista)
```

# **Cap√≠tulo 5 - Fun√ß√µes Auxiliares √öteis**

Conta o n√∫mero de elementos em uma lista encadeada.

```python
def contar_nos(lista):
    contador = 0
    atual = lista
    while atual is not None:
        contador += 1
        atual = atual["prox"] if "prox" in atual else atual["proximo"]
    return contador

# Teste
no3 = {"valor": 30, "prox": None}
no2 = {"valor": 20, "prox": no3}
no1 = {"valor": 10, "prox": no2}
lista_teste = no1

print("‚úÖ Teste contar_nos:", contar_nos(lista_teste))  # Sa√≠da esperada: 3
```

Converte uma lista encadeada para uma lista Python comum.

```python
def para_lista_python(lista):
    resultado = []
    atual = lista
    while atual is not None:
        resultado.append(atual["valor"])
        atual = atual["prox"] if "prox" in atual else atual["proximo"]
    return resultado

# Teste
print("‚úÖ Teste para_lista_python:", para_lista_python(lista_teste))  # Sa√≠da esperada: [10, 20, 30]
```

Verifica se um valor existe na lista encadeada.

```python
def buscar(lista, valor):
    atual = lista
    while atual is not None:
        if atual["valor"] == valor:
            return True
        atual = atual["prox"] if "prox" in atual else atual["proximo"]
    return False

# Teste
print("‚úÖ Teste buscar (existe):", buscar(lista_teste, 20))  # Sa√≠da esperada: True
print("‚úÖ Teste buscar (n√£o existe):", buscar(lista_teste, 99))  # Sa√≠da esperada: False
```
